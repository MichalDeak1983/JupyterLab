Otra cosa que choca es que la columna "died" tenga valores decimales. No tiene mucho sentido, ya que se trata de año. Se trata de un problema con los valores np.NaN, que se interpretan como valores decimales, por lo que pandas convierte todos los valores de la columna en float. Para solucionar el problema, pandas nos ofrece otro tipo de datos ('Int64', con mayúscula'), que convierte los valores np.NaN en pandas.NA


### Ejercicio 6
Convierte el tipo de dato de la columna 'died' y 'born' a Int64

```python
# Resuelve aquí
# Tipo podemos cambiar con metodo astype(nombre_tipo)
autores['born']=autores['born'].astype('int64')
```

```python
# Para cambiar la columna con valores nulos a tipo int64 tenemos asignar valores numericos a los nulos
# ya que todos anos de nacimiento son mayores de 1000 he replazado los 'NaN' con 0s, en esta manera 
# sabemos que el valor 0 de ano no tiene sentido porque en  ano del muerte es menor de ano de nacimiento
autores['died']=autores['died'].replace(np.nan,0).astype('int64')
```

### Ejercicio 15

- ¿Cuál es la media de edad de mis mis autores, en general? Para ello, crea una columna nueva con la edad aproximada de los autores. (Una pista: utiliza como referencia el año 2021)
- ¿Cuál es la media de edad de los autores muertos?
- ¿Cuál es la media de edad de los autores vivos?
- ¿Y la media de edad a nivel de género?

```python
# Resuelve aquí
# Para calcular edad replazamos los 0s en anos de muerte en autores con 2021 y sustraemos anos de nacimiento
autores['age']=autores['died'].replace(0,2021)-autores['born']
# calculamos edad media
print('Edad media de los autores:',autores['age'].mean())
# Lo mismo con solo los autores vivos:
print('Edad media de los autores vivos:',autores['age'][~autores['dead']].mean())
# Lo mismo con solo los autores muertos:
print('Edad media de los autores muertos:',autores['age'][autores['dead']].mean())
```

```python
wiki = wikipedia.page('El tambor de hojalata (novela)')
```

```python
text = wiki.content
```

```python
text
```

```python
text = re.sub(r'==.*?==+', '', text) # eliminamos los headers
text = text.replace('\n', ' ') # eliminamos los saltos de línea
""" Un cambio pequenito arriba: he replazado '' con ' ' porque si en que sique se usa .split() mejor replazar '\n' con un espacio. No se que exactamente hace WordCloud. """
```

```python
# Creamos una función para generar el gráfico
def plot_cloud(wordcloud):
    # fijamos el tamaño
    plt.figure(figsize=(40, 30))
    # plot de la imagen
    plt.imshow(wordcloud) 
    # sin ejes
    plt.axis("off");
```

```python
# Importamos el paquete para hacer la nube de palabras
from wordcloud import WordCloud
```

```python
#Genaramos la nube
wordcloud = WordCloud(width = 3000, height = 2000, random_state=1, max_words=50,
                      background_color='salmon',colormap='Pastel1', 
                      collocations=False, stopwords = stop_words).generate(text)
```

```python
plot_cloud(wordcloud)
```
